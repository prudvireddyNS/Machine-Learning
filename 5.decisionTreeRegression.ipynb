## Data Preprocessing

```python
# changing datatypes
cleaned_data['tpep_dropoff_datetime'] = pd.to_datetime(cleaned_data['tpep_dropoff_datetime'])
cleaned_data['tpep_pickup_datetime'] = pd.to_datetime(cleaned_data['tpep_pickup_datetime'])

# Extracting hours from dropoff and pickup
cleaned_data['pickup_hour'] = cleaned_data['tpep_pickup_datetime'].dt.hour
cleaned_data['dropoff_hour'] = cleaned_data['tpep_dropoff_datetime'].dt.hour

# Extracting day of week from dropoff and pickup
cleaned_data['pickup_day'] = cleaned_data['tpep_pickup_datetime'].dt.weekday
cleaned_data['dropoff_day'] = cleaned_data['tpep_dropoff_datetime'].dt.weekday

# Calculating trip time in seconds
cleaned_data['trip_time'] = (cleaned_data['tpep_dropoff_datetime'] - cleaned_data['tpep_pickup_datetime']).dt.total_seconds()

# Ideally use the full dataset for this exercise.
# However, if you run into out-of-memory issues due to the data size, reduce it.
# For instance, in this example, we use only the first 200,000 samples.
rows = 2000000
cleaned_data = cleaned_data.head(rows)
```

```python
# drop pickup and dropoff datetimes
cleaned_data = cleaned_data.drop(['tpep_dropoff_datetime','tpep_pickup_datetime'],axis=1)

# encoding the categorical features
get_dummy_col = ["VendorID","RatecodeID","store_and_fwd_flag","PULocationID", "DOLocationID","payment_type", "pickup_hour", "dropoff_hour", "pickup_day", "dropoff_day"]
proc_data = pd.get_dummies(cleaned_data, columns = get_dummy_col)

# deleting cleaned_data
del cleaned_data
gc.collect()
```

```python
# extract target variable
y = proc_data['tip_amount']
# drop target variable
proc_data = proc_data.drop('tip_amount',axis=1)

# get the feature matrix used for training
X = proc_data.values

# normalize the feature matrix
from sklearn.preprocessing import normalize

X = normalize(X,axis=1, norm='l1', copy = False)

X.shape, y.shape
```

```python
proc_data
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# building Regression tree model from scikit-learn
from sklearn.tree import DecisionTreeRegressor
import time

tree = DecisionTreeRegressor(max_depth=8, random_state=35)

# training the model
t = time.time()
tree.fit(X_train, y_train)
t1 = time.time() - t
print("Time taken to train is ",t1)
```
